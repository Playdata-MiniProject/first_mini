```python
import pandas as pd
```

```python
# 종목시세정보 파일 불러오기

stock_price = pd.read_csv('./data/NASDAQ_DT_FC_STK_QUT.csv')
stock_price.head()
```

```python
# 종목시세정보 5월~8월

data = stock_price[222183:]
data
```

```python
# 유니크한 값 확인

print(data['trd_dt'].unique())
print(data['gts_iem_ong_pr'].unique())
print(data['gts_iem_hi_pr'].unique())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib import font_manager, rc


# 시가 종가 기울기 계산
data['price_slope'] = (data['gts_iem_end_pr'] - data['gts_iem_ong_pr']) / data['gts_iem_ong_pr']


# 표준화 및 정규화
data['price_slope_standardized'] = (data['price_slope'] - data['price_slope'].mean()) / data['price_slope'].std()
data['price_slope_normalized'] = (data['price_slope_standardized'] - data['price_slope_standardized'].min()) / (data['price_slope_standardized'].max() - data['price_slope_standardized'].min())


# 이상치 제거 (20% 이하)
z_scores = np.abs(stats.zscore(data['price_slope_normalized']))
data_no_outliers = data[z_scores <= 0.2]


# 한글 폰트 설정
font_path = r'C:\Windows\Fonts\malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)


# 정규분포 그리기
plt.hist(data_no_outliers['price_slope_normalized'], bins=10, density=True, alpha=0.6, color='g')
plt.title('정규화된 주가 기울기 분포')
plt.xlabel('정규화된 주가 기울기')
plt.ylabel('밀도')
plt.show()
```
