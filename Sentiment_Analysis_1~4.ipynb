# 감성분석

### 데이터 불러오기

```python
import pandas as pd


# RSS_total 데이터 불러오기
rss_info = pd.read_csv('./data/RSS_total.csv', encoding='ANSI')

# 1월~4월 데이터 슬라이싱
data = rss_info[:15077]

# Unnamed 열 삭제
data.drop(columns=data.columns[data.columns.str.startswith('Unnamed')], inplace=True)

# 인덱스 0부터 시작하도록 리셋
data.reset_index(drop=True, inplace=True)

# 확인
data
```

### 감성분석

```python
from textblob import TextBlob


# "news_smy_ifo" 열의 감성분석 수행
sentiments = []
for text in data['news_smy_ifo']:
    analysis = TextBlob(text)
    sentiment = analysis.sentiment.polarity
    sentiments.append(sentiment)
    

# 감성분석 결과를 새로운 열로 추가
data['sentiment'] = sentiments


# 결과 출력
data
```

```python
# 파일로 저장

data.to_csv('./data/Summary_Sentiment.csv', encoding='ANSI', index=False)
```
