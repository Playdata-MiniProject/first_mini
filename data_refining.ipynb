## 데이터 정제

```python
import pandas as pd
import os
```

## 뉴스 데이터

### 병합

```python
# 뉴스데이터 파일 리스트
csv_list = os.listdir('./data/NASDAQ_RSS_IFO')
csv_list
```

```python
# 파일들의 path룰 리스트에 저장
RSS_list = [
    pd.read_csv('./data/NASDAQ_RSS_IFO/'+y, encoding='ANSI')
    for y in csv_list
]
```

```python
# 리스트에 잘 들어갔는지 확인하기
RSS_list[0].iloc[0]
```

```python
# 리스트 속 path를 이용해 세로 방향으로 병합
RSS = pd.concat(RSS_list)
RSS.head()
```

### 병합으로 섞인 index 리셋하기

```python
RSS.reset_index(inplace=True)
RSS.drop('index',inplace=True,axis=1)
```

```python
RSS
```

```python
RSS.info()
```

### 결측치 확인

```python
RSS.isnull().sum()
```

### 중복제거

```python
# 중복값 있는지 확인
print(len(RSS))
print(len(RSS.drop_duplicates()))
```

```python
# 중복제거
RSS.drop_duplicates(inplace=True)
RSS.head()
```

```python
# 인덱스 리셋
RSS.reset_index(inplace=True)
```

### rld_ose_iem_tck_cd 컬럼에서 중복된 값을 갖는 열을 찾아 재정의

```python
# 실험
print(RSS.rld_ose_iem_tck_cd[146913].split(','))
print(','.join(set(RSS.rld_ose_iem_tck_cd[146913].split(','))))
```

```python
# 중복된 값을 갖는 열을 찾아 재정의
for i in range(len(RSS)):
    split_result = RSS.rld_ose_iem_tck_cd[i].split(',')
    if len(split_result) >= 2:
        RSS.rld_ose_iem_tck_cd[i] = ','.join(set(split_result))
```

```python
# 확인
RSS
```

```python
# 아까 인덱스 리셋하면서 생긴 index 컬럼 제거
RSS.drop('index',axis=1,inplace=True)
```

```python
RSS
```

```python
# 파일로 저징
RSS.to_csv('./data/RSS_total.csv', encoding='ANSI')
```
